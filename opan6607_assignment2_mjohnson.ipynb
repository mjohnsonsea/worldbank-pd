#### Q23
Create a new column that is the log of the gdpPercap column and create a scatter plot (with matplotlib, seaborn, or altair) with log of gdp on the x-axis and life expectancy on the y-axis. Drawing on this plot and the result from #25, write a paragraph explaining your results and some potential explanations for the relationship or lack thereof.

```python
# Create new column that is the log of gdpPercap
df['gdpPercap_log'] = np.log(df['gdpPercap'])
```

```python
# Import seaborn
import seaborn as sns
```

```python
# Create scatterplot of gdpPercap_log and lifeExp
q23_scatterplot = sns.scatterplot(x = 'gdpPercap_log', 
                                  y = 'lifeExp',
                                  data = df)

# Add labels and title
q23_scatterplot.set(xlabel = 'Log of GDP per Capita',
                    ylabel = 'Life Expectancy',
                    title = 'Log of GDP per Capita vs Life Expectancy')
```

The scatterplot shows a positive correlation between GDP per capita (on a logarithmic scale) and life expectancy. This means that, generally, countries with a higher GDP per capita tend to have higher life expectancies. The visuals in Q24 and Q25 show regional differences. For example, North American and European countries cluster toward the higher end of the both axes. On the other hand, African countries are clustered toward the lower end of both axes.

Create another scatterplot (with matplotlib, seaborn, or altair) with log of gdp and life expectancy and color the points by continent.

#### Q25
Create a scatterplot with altair with log of gdp and life expectancy, color the points by continent, add tooltip functionality so that hovering over the data points will show the continent, country, and year of that data point, and add a title.

```python
# Import altair
import altair as alt
```

```python
# Create scatterplot
q25_scatterplot = alt.Chart(df).mark_circle().encode(
    x = alt.X('gdpPercap_log', title = 'Log of GDP per Capita'),
    y = alt.Y('lifeExp', title = 'Life Expectancy'),
    color = alt.Color('continent', title = 'Continent'),
    tooltip = ['continent', 'country', 'year']).properties(
    title = 'Log of GDP per Capita vs Life Expectancy by Continent')

q25_scatterplot
```
